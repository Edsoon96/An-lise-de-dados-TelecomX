```python
import pandas as pd
import requests

# URL do arquivo JSON no GitHub
url = "https://raw.githubusercontent.com/ingridcristh/challenge2-data-science/main/TelecomX_Data.json"

# Requisição à API
response = requests.get(url)
data = response.json()

# Conversão para DataFrame
df = pd.json_normalize(data)

# Visualização inicial
df.head()
```

```python
df["account.Charges.Total"] = pd.to_numeric(
    df["account.Charges.Total"], errors="coerce"
)
```

```python
# Exemplos de colunas categóricas
pd.unique(df["account.Contract"])
pd.unique(df["internet.InternetService"])
pd.unique(df["phone.MultipleLines"])
```

```python
import numpy as np


df["Churn"] = df["Churn"].replace("", np.nan)


df["Churn"].isna().sum()
```

```python
df = df.dropna(subset=["Churn"])
```

```python
df["account.Charges.Total"] = pd.to_numeric(
    df["account.Charges.Total"], errors="coerce"
)
```

```python
df["account.Charges.Total"].isna().sum()
```

```python
colunas_categoricas = [
    "customer.gender",
    "account.Contract",
    "account.PaymentMethod",
    "internet.InternetService"
]

for col in colunas_categoricas:
    df[col] = df[col].str.lower().str.strip()
```

```python
df["internet.InternetService"] = df["internet.InternetService"].replace(
    "no internet service", "no"
)
```

```python
df.info()
```

```python
df["Contas_Diarias"] = df["account.Charges.Monthly"] / 30
```

```python
df[["account.Charges.Monthly", "Contas_Diarias"]].head()
```

```python
df["Churn"] = df["Churn"].map({"Yes": 1, "No": 0})
```

```python
colunas_binarias = [
    "phone.PhoneService",
    "phone.MultipleLines",
    "internet.OnlineSecurity",
    "internet.OnlineBackup",
    "internet.DeviceProtection",
    "internet.TechSupport",
    "internet.StreamingTV",
    "internet.StreamingMovies",
    "account.PaperlessBilling"
]

for col in colunas_binarias:
    df[col] = df[col].replace({"Yes": 1, "No": 0})
```

```python
df[colunas_binarias] = df[colunas_binarias].replace(
    {"No internet service": 0, "No phone service": 0}
)
```

```python
df = df.rename(columns={
    "customer.gender": "Genero",
    "customer.SeniorCitizen": "Idoso",
    "customer.Partner": "Possui_Parceiro",
    "customer.Dependents": "Possui_Dependentes",
    "customer.tenure": "Tempo_Contrato",
    "account.Charges.Monthly": "Cobranca_Mensal",
    "account.Charges.Total": "Cobranca_Total"
})
```

```python
df["Tempo_Contrato"] = df["Tempo_Contrato"].astype(int)
df["Cobranca_Mensal"] = df["Cobranca_Mensal"].astype(float)
df["Cobranca_Total"] = df["Cobranca_Total"].astype(float)
```

Neste trabalho, o objetivo foi entender melhor o problema de evasão de clientes da Telecom X, tentando identificar, de forma simples e prática, os principais motivos que levam os clientes a cancelarem os serviços. Ao longo da análise, os dados foram importados da API, organizados e tratados com cuidado, corrigindo inconsistências, ajustando formatos e criando novas informações, como o valor de cobrança diária, para facilitar a interpretação. A partir disso, foram feitas análises descritivas e exploratórias, com gráficos e comparações entre clientes que permaneceram e os que evadiram, o que mostrou claramente que clientes com pouco tempo de contrato, contratos mensais e valores mais altos de cobrança tendem a cancelar com mais frequência. Também foi possível perceber que clientes que contratam mais serviços costumam permanecer por mais tempo. Com base nesses achados, fica evidente que a empresa pode reduzir a evasão investindo em ações de fidelização logo no início do contrato, oferecendo incentivos para planos mais longos e criando estratégias que estimulem o uso de mais serviços, fortalecendo o relacionamento com o cliente e aumentando sua satisfação.
